```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('/content/drive/MyDrive/CDK-HackaThon/Credit data report 2 (1).xlsx',index_col=False)
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(['Monthly Expenses','Monthly Debts','Default eligible amount','Employment Status','Customer ID','First Name', 'Middle Name', 'Date of Birth','Last Name', 'Email', 'Home Phone', 'Address Line1','postalZip','city','State'], axis = 1,inplace = True)
```

```python
df.head()
```

```python
df.drop(['Decision'],inplace=True,axis=1)
```

```python
print(df['Red Flag'].unique())
print(df['Default in Last 7 days'].unique())
print(df['Month1'].unique())
print(df['Month2'].unique())
print(df['Month3'].unique())
```

```python
df.info()
```

```python
df.drop(['Total Outstanding'],inplace=True,axis=1)
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
y=df['Eligible amount']
x=df.drop(['Eligible amount'],axis=1)
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.isna().sum()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
n_estimators = [int(x) for x in np.linspace(start = 1, stop = 12, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(3, 15, num = 6)]
min_samples_split = [2, 3,5, 7,10,]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
y_predRf=rf_random.predict(x_test)
```

```python
r2_score(y_test,y_predRf)
```
